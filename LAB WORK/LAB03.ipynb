```python
def functions(n):
    sum=0
    length=0
    for i in range(len(n)): 
        sum=sum+(n[i]) #used to add every element stored in the list cumulatively.
    for i in n:
        length=length+1 #as it encounters an element in the list, the value of length increments by one. It hence calculates the length of the given list, which in our case is the number of villagers
    print("Total sum is", sum)
    print("Length is", length)
    print("Mean value is total sum/length, that is:", sum/length,"\n") #mean value is calculated

    print("*"*50)
    print("MINIMUM:")

    minimum=n[0] #the first element of the set is assumed to be the minimum.
    for i in range(length): #the assumed value is compared to every other element using the for loop.
        if minimum<n[i]: #if the number the minimum value is being compared with is less than the minimum value,
            minimum=minimum #the minimum value is replaced with the lower one. The process repeats until every element in the list is compared.
        else:
            minimum=n[i]
    print("The minimum value is:", minimum)
    print("The minimum value using in-built functions is:",min(n),"\n") #comparing the output with the in-built function.

    print("*"*50)
    print("MAXIMUM:")

    maximum=n[0] #the first element of the set is assumed to be the maximum
    for i in range(length): #the assumed value is compared to every other element using the for loop.
        if maximum>n[i]: #if the number the maximum value is being compared with is greater than the maximum value,
            maximum=maximum #the maximum value is replaced with the greater one. The process repeats until every element in the list is compared.
        else:
            maximum=n[i] #if the condition is true the value stored remains.
    print("The maximum value is:", maximum)
    print("The maximum value using in-built functions is:",max(n)) #comparing the output with the in-built function.
    
    print("*"*50)
    print("RANGE:")

    ranges=maximum-minimum #the range of the list is calculated
    print("The range is:",ranges)

    return True

functions(villagers)
```
