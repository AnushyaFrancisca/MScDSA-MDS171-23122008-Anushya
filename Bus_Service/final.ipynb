```python
oldid=[] #stores all the passenger id's generated
passenger_details={} #an empty dictionary to store passenger details

#the following function generates a unique five digit ID for each user

def Passenger_details(bs_no):
    passengerid=str(random.randint(10000,99999))
    oldid.append(passengerid)
    unique(passengerid)
    storeDetails(passengerid,bs_no)
    print("Your passenger ID is: ",passengerid)

#the following function is used to check whether the passenger ID is unique or not. If the passenger ID already exists, a new ID is generated again.

def unique(passengerid):
    if passengerid in oldid:
        passengerid=str(random.randint(10000,99999))
        oldid.append(passengerid)
    else:
        oldid.append(passengerid)

#the function is defined to store the details of the passenger who is availing the services offered.

def storeDetails(passengerid,bs_no):
    choice=input("Enter your details in order to proceed with the booking. If you wish to continue enter 'Y' else enter 'N'.").title()
    if choice=='Y':
        name=input("Enter the name of the student.").strip().title()
        reg=input("Enter the registration number of the student.")
        phone=input("Enter the mobile number of the student.")
        email=input("Enter the email id of the student.")
        file.write("-"*30+"\n")
        file.write("STUDENT DETAILS\n") #details to be stored in the csv file
        file.write("PASSENGER ID: "+passengerid+"\n") #details to be stored in the csv file
        file.write("NAME: "+name+"\n") #details to be stored in the csv file
        file.write("REG. NO:"+reg+"\n") #details to be stored in the csv file
        file.write("CONTACT: "+phone+"\n") #details to be stored in the csv file
        file.write("EMAIL ID: "+email+"\n") #details to be stored in the csv file
        file.write("Seat/Seats Booked: \n")
        for i in bs_no:
            file.write(i+" ")
        file.write("\n")
        file.write("-"*45) #details to be stored in the csv file
        passenger={ #stores the details of the passenger in a dictionary
            "name":name,
            "reg. no":reg,
            "phone":phone,
            "email":email,
            "seat no.":bs_no
        }
        passenger_details[passengerid]=passenger #appends the dictionary to a main dictionary that stores the passenger ID as the key and the passenger details as the value
    else:
        print("You have chosen to exit.")
        exit()
    
```

```python
#this function is used to view the available bus seats

def view_bus(as_no,Bs_no):
    columns=2 #used to specify that the list is to be taken in pairs of two
    #zip is used to comnine both the lists created below 
    for first, second in zip(as_no[::columns] , as_no[1::columns]): #[::columns] creates a new list containing 1st, 3rd... elements, [1::columns] creates a new list containing 2nd, 4th... elements
            if str(first) in Bs_no and str(second) in Bs_no: #if the seat booked is in the first and second column, it prints both in red
                print(f'{Fore.RED}{first: <10}{Fore.RED}{second}')  
                print(Style.RESET_ALL)
            elif str(first) in Bs_no: #if the seat booked is in the first column, it prints in red
                print(f'{Fore.RED}{first: <10}{Fore.GREEN}{second}')
                print(Style.RESET_ALL)    
            elif str(second) in Bs_no: #if the seat booked is in the second column, it prints in red
                print(f'{Fore.GREEN}{first: <10}{Fore.RED}{second}')
                print(Style.RESET_ALL)
            else:
                print(f'{Fore.GREEN}{first: <10}{Fore.GREEN}{second}') #if the seat booked is in neither of the columns of the pair, it prints both in green
                print(Style.RESET_ALL)
```

```python
#this function is used to view seats after booking

def booked(as_no,Bs_no):
        columns=2 
        for first, second in zip(as_no[::columns] , as_no[1::columns]):
            if str(first) in Bs_no and str(second) in Bs_no: #if the seat booked is in the first and second column, it prints both in red
                print(f'{Fore.RED}{first: <10}{Fore.RED}{second}')  
                print(Style.RESET_ALL)
            elif str(first) in Bs_no: #if the seat booked is in the first column, it prints in red
                print(f'{Fore.RED}{first: <10}{Fore.GREEN}{second}')
                print(Style.RESET_ALL)    
            elif str(second) in Bs_no: #if the seat booked is in the second column, it prints in red
                print(f'{Fore.GREEN}{first: <10}{Fore.RED}{second}')
                print(Style.RESET_ALL)
            else:
                print(f'{Fore.GREEN}{first: <10}{Fore.GREEN}{second}') #if the seat booked is in neither of the columns of the pair, it prints both in green
                print(Style.RESET_ALL)
```

```python
#function used for booking a seat

def seat_no(bs_no,Bs_no):
        number=int(input("Enter the no of seats you want to book: "))
        for i in range(number):
            seat_number=input("Enter the seat number that you want to book: ")
            if seat_number in Bs_no: #condition to check if the code is already booked or not
                print("Seat already booked by someone else, please try booking another seat.")
            else:
                bs_no.append(seat_number)
                Bs_no.append(seat_number)
                pass
        bs_no.sort()
```

```python
#the bus pass of the passenger is displayed when the passenger enters the ID provided to them
#library imported to format output in the form of tables

def displayDetails(passenger):
    if passenger in passenger_details:
        print("Your bus pass is displayed below:")
        t = PrettyTable([ 'Name', 'Reg. No', 'Phone', 'Email ID'])  #four columns with the heading as mentioned
        t.add_row([passenger_details[passenger]["name"], passenger_details[passenger]["reg. no"], passenger_details[passenger]["phone"], passenger_details[passenger]["email"]])
        print(t)
        print("Seats booked: ",passenger_details[passenger]["seat no."])
    else:
        print(Fore.RED+"Passenger ID does not exist.")
    
```

```python
#the below function permits the user to cancel a seat that has been booked

def cancellation(passenger):
    if passenger in passenger_details:
        bus_number=input("Enter the serial number of the bus you have booked a seat in.")

        if bus_number=='1':
            delete=[] #empty list to store the seat number which needs to be cancelled
            n=int(input("Enter the total number of seats you want to cancel."))

            for i in range(n):
                n1=input("Enter the seat number you want to delete:")
                delete.append(n1) 

            for i in delete: 
                if i in Bs1: #if the seat no. given by the user is present in the booked seats list then the specific seat is removed from the booked list
                    Bs1.remove(i)

            print("Your cancellation has been processed. The available seats are: ")
            booked(as1,Bs1)

        elif bus_number=='2':
            delete=[] #stores the seat number which needs to be cancelled
            n=int(input("Enter the total number of seats you want to cancel."))
            for i in range(n):
                n1=input("Enter the seat number you want to delete:")
                delete.append(n1)

            for i in delete:
                if i in Bs2:
                    Bs2.remove(i)

            print("Your cancellation has been processed. The available seats are: ")
            booked(as2,Bs2)

        elif bus_number=='3':
            delete=[] #stores the seat number which needs to be cancelled
            n=int(input("Enter the total number of seats you want to cancel."))
            for i in range(n):
                n1=input("Enter the seat number you want to delete:")
                delete.append(n1)

            for i in delete:
                if i in Bs3:
                    Bs3.remove(i)

            print("Your cancellation has been processed. The available seats are: ")
            booked(as3,Bs3)

        elif bus_number=='4':
            delete=[] #stores the seat number which needs to be cancelled
            n=int(input("Enter the total number of seats you want to cancel."))
            for i in range(n):
                n1=input("Enter the seat number you want to delete:")
                delete.append(n1)

            for i in delete:
                if i in Bs4:
                    Bs4.remove(i)

            print("Your cancellation has been processed. The available seats are: ")
            booked(as4,Bs4)

        else:
            print(Fore.RED+"Invalid input")
            
    else:
        print(Fore.RED+"Passenger ID does not exist. Please try again.")
```
