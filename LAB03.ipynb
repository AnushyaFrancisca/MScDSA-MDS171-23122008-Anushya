```python
sum=0
length=0
for i in range(len(villagers)):
    sum=sum+(villagers[i])
for i in villagers:
    length=length+1
print("Total sum is", sum)
print("Length is", length)
print("Mean value is total sum/length, that is:", sum/length) 
```

```python
minimum=villagers[0]
for i in range(length):
    if minimum<villagers[i]:
        minimum=minimum
    else:
        minimum=villagers[i]
print("The minimum value is:", minimum)
print("The minimum value using in-built functions is:",min(villagers))
```

```python
maximum=villagers[0]
for i in range(length):
    if maximum>villagers[i]:
        maximum=maximum
    else:
        maximum=villagers[i]
print("The maximum value is:", maximum)
print("The maximum value using in-built functions is:",max(villagers))
```

```python
range=maximum-minimum
print("The range is:",range)
```
